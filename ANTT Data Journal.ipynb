A calculadora de amostragem diz que, para a quantidade de observações no conjunto de dados, eu preciso de uma amostragem de pelo menos 3392 observações para uma margem de erro de 2% e nível de confiança de 95%. Vou usar 15% das observações totais, que será em torno de 1,6 milhão.

Só que eu quero uma amostragem que me permita representar as saídas de todo o país e todos os dois anos de forma uniforme. Para isso, farei uma amostragem estratificada.

```python
# Organizando os dados por estado...

df["ponto_origem_viagem"] = df["ponto_origem_viagem"].astype(str)
df["ponto_origem_viagem"] = df["ponto_origem_viagem"].str.strip()
```

```python
df["uf_origem"] = df["ponto_origem_viagem"].apply(lambda x: x[-3:])
df["uf_origem"].unique()
```

```python
df["uf_origem"] = df["uf_origem"].str.replace("/", "")
df["uf_origem"].unique()
```

```python
# Criando a amostra estratificada.
# Primeiro separamos o conjunto original por ano e estado, depois capturamos o equivalente a 15% de cada categoria.

stratified_sample = df.groupby(["ano", "uf_origem"]).apply(lambda x: x.sample(frac=0.15, replace=False, random_state=666))

stratified_sample = stratified_sample.droplevel(0)
stratified_sample = stratified_sample.droplevel(0)
stratified_sample
```

```python
# Verificando a proporção por ano...

stratified_sample["ano"].value_counts(normalize=True)
```

```python
# Verificando a proporção por estado...

stratified_sample["uf_origem"].value_counts(normalize=True)
```

```python
# Conferindo a proporção original mais uma vez...

df["uf_origem"].value_counts(normalize=True)
```

```python
# Uma última checagem nas colunas...

stratified_sample.info()
```

Com tudo isso feito, fica mais fácil fazer algumas limpezas adicionais.

```python
# Verificando se existem valores negativos...

stratified_sample[stratified_sample["valor_tarifa"] < 0]
```

```python
# Passando tudo para o valor absoluto...

stratified_sample["valor_tarifa"] = abs(stratified_sample["valor_tarifa"])
stratified_sample[stratified_sample["valor_tarifa"] < 0]
```

```python
stratified_sample["valor_percentual_desconto"] = abs(stratified_sample["valor_percentual_desconto"])
stratified_sample["valor_aliquota_icms"] = abs(stratified_sample["valor_aliquota_icms"])
stratified_sample["valor_pedagio"] = abs(stratified_sample["valor_pedagio"])
stratified_sample["valor_taxa_embarque"] = abs(stratified_sample["valor_taxa_embarque"])
stratified_sample["valor_total"] = abs(stratified_sample["valor_total"])
```
